```python
import torch
from torch.nn import functional as F
import matplotlib.pyplot as plt
import numpy as np
```

```python
#x = list(range(10))
x = torch.linspace(-1000,1000,100).cpu()
y = F.logsigmoid(x)
yy = F.logsigmoid(-x)
plt.plot(x.numpy(),y.numpy())
#plt.plot(x.numpy(),yy.numpy())
y[99]
```

```python
#x = list(range(10))
x = torch.linspace(-1000,1000,100).cpu()
y = torch.sigmoid(x)
yy = torch.sigmoid(-x)
plt.plot(x.numpy(),y.numpy())
#plt.plot(x.numpy(),yy.numpy())
y[99]
```
